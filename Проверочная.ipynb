```python
import pandas as pd
from datetime import datetime
```

```python
users = pd.read_excel("users.xlsx") # загружаем данные по клиентам
payments = pd.read_excel("payments.xlsx") # загружаем данные по платежам
workouts = pd.read_csv("workouts.csv") # загружаем данные по тренировкам
```

## Задание 1

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
payments['payment_date'] = pd.to_datetime(payments['payment_date'], errors='coerce')
```

```python
payments
```

```python
payments2020 = payments.loc[payments['payment_date'].dt.year == 2020]
```

```python
payments2020
```

```python
USD_RATE = 73.14 # курс доллара

def map_currency(currency): # определи функцию для заполнения столбца fx_rate методом .apply()
  if currency == 'USD':
    return USD_RATE
  else:
    return 1
```

```python
payments2020['fx_rate'] = payments2020['currency']
```

```python
payments2020['fx_rate'] = payments2020['fx_rate'].apply(map_currency)
```

```python
payments2020['amount_rub'] = payments2020['amount'] * payments2020['fx_rate']
```

```python
payments2020
```

```python
pay_month = payments2020['payment_date'].dt.month
```

```python
payments2020.groupby([payments2020['payment_date'].dt.month, 'workout_type'])['amount_rub'].sum()
```

Вывод: general тренировки приносят много денег, может они смогут приносить еще больше

```python
user_regions = users[['user_id', 'region']]
```

```python
use_and_pay = pd.merge(user_regions, user_payments,
               how = 'left',
               left_on = 'user_id', right_on = 'user_id')
```

```python
usepay = use_and_pay.dropna()
```

```python
usable = usepay.groupby(usepay['region'])['amount_rub'].sum().reset_index().sort_values(by='amount_rub').reset_index()
```
