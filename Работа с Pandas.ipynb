```python
import pandas as pd
```

```python
medc = pd.read_excel("medication.xlsx")
move = pd.read_excel("movement.xlsx")
phrm = pd.read_excel("pharmacies.xlsx")
```

```python
medc
```

```python
move
```

```python
phrm
```

```python
new = pd.merge(move, phrm,
               how = 'left',
               left_on = 'ID аптеки', right_on = 'ID аптеки')
```

```python
new
```

```python
final = pd.merge(new, medc,
               how = 'left',
               left_on = 'Артикул', right_on = 'Артикул')
```

```python
final
```

```python
final['Стоимость'] = final['Количество упаковок, шт.'] * final['Цена за упаковку, руб.']
```

```python
final
```

## 1. Название самого дешевого лекарства

```python
best_price = min(final['Цена за упаковку, руб.'])
```

```python
name_of_the_best_price = final.loc[final['Цена за упаковку, руб.'] == best_price]['Название'].reset_index().loc[0]
```

```python
name_of_the_best_price
```

## 2. Название сетей аптек с адресами где оно продается

```python
name_webs = final.loc[final['Цена за упаковку, руб.'] == best_price][['Сеть', 'Адрес']]
```

```python
name_webs
```

## 3. Самое дешевое обезболивающее (анальгетик) в сети ФармСтандарт (что, где, сколько)

```python
anal = final[['Название', 'Сеть', 'Адрес', 'Группа препаратов', 'Цена за упаковку, руб.']].loc[final['Группа препаратов'] == 'анальгетик']
```

```python
best_price_anal = min(anal['Цена за упаковку, руб.'])
```

```python
anal.loc[(anal['Цена за упаковку, руб.'] == best_price_anal) & (anal['Сеть'] == 'ФармСтандарт')][['Название', 'Адрес', 'Цена за упаковку, руб.']]
```
